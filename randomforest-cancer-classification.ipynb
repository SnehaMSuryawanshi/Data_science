```python
categories=['Race', 'Marital Status', 'T Stage ', 'N Stage', '6th Stage',
       'differentiate', 'Grade', 'A Stage', 'Estrogen Status',
       'Progesterone Status']
```

```python
x=df.drop('Status',axis=1)
y=df['Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in categories:
    x_train[i]=l.fit_transform(x_train[i])
    x_test[i]=l.transform(x_test[i])
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.get_params()
```

```python
param_grid = {
    'n_estimators': [25, 50, 100, 150],
    'max_features': ['sqrt', 'log2', None],
    'max_depth': [3, 6, 9],
    'max_leaf_nodes': [3, 6, 9],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(model,param_grid)
random_search.fit(x_train, y_train)
print(random_search.best_estimator_)
```

```python
from sklearn.ensemble import RandomForestClassifier
h=RandomForestClassifier(n_estimators=100,random_state=1)
```

```python
h.fit(x_train,y_train)
```

```python
y_pred=h.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
